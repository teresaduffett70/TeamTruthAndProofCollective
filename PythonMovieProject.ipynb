## Clean dataset up

```python
movies['GrossR'] = movies['Gross'].str.replace(',','')
```

```python
movies['GrossR'] = movies['GrossR'].str.replace("'", "")
```

```python
movies['GrossR']=movies['GrossR'].fillna(0)
```

```python
movies['GrossR']=movies['GrossR'].dropna(0)
```

```python
movies['GrossR'] = movies['GrossR'].astype(int)
```
